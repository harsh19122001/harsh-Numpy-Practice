# Array Manipulation functions

```python
# Reshapes an array
a1 = np.array([1,2,3])
reshaped = np.reshape(a1 , (1,3)) # Reshape to 1x3
print("Reshaped array : ",reshaped)
```

```python
# Reshapes an array
a1 = np.array([1,2,3])
reshaped = np.reshape(a1 , (3,1)) # Reshape to 1x3
print("Reshaped array : ",reshaped)
```

```python
# Flatten an array
f1 = np.array([[1,2],[3,4]])
flattened = np.ravel(f1) #Flatten to 1D array
print("Flattened array:", flattened)
```

```python
# Transpose an array
e1 = np.array([[1,2],[3,4]])
transposed = np.transpose(e1) # Transpose the array
print("Transposed array : \n",transposed)
```

```python
# Stack array vertically
a2 = np.array([1 , 2])
b2 = np.array([3 , 4])
stacked = np.vstack([a2 , b2]) #stack a and b vertically
print("Stacked array: \n", stacked)
```

# Broadcasting

```python
# same shape
a = np.arange(6).reshape(2,3)
b = np.arange(6,12).reshape(2,3)

print(a)
print(b)
```

```python
print(a+b)
```

```python
# diff shape
a = np.arange(6).reshape(2,3)
b = np.arange(3).reshape(1,3)

print(a)
print(b)
```

```python
print(a+b)
```

```python
a = np.arange(12).reshape(4,3)
b = np.arange(3)

print(a) 
print('\n',b)
```

```python
print(a+b) # Arithmetic Operation
```

```python
# could not Broadcast
a = np.arange(12).reshape(3,4)
b = np.arange(3) 

print(a)
print('\n',b)
```

```python
print(a+b) # could not broadcast
```

```python
a = np.arange(3).reshape(1,3)
b = np.arange(3).reshape(3,1)

print(a) 
print('\n',b)
```

```python
print(a+b)
```

```python
a = np.arange(3).reshape(1,3)
b = np.arange(4).reshape(4,1) 

print(a)
print('\n',b)
```
